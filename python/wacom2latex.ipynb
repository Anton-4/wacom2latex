```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import train_test_split
from  sklearn.preprocessing import MinMaxScaler

def split_data(df, test_perc):
    train, test = train_test_split(df, test_size = test_perc)

    x_train = train.iloc[:,:-1].as_matrix()
    y_train = train[['label']].values.flatten()

    x_test = test.iloc[:,:-1].as_matrix()
    y_test = test[['label']].values.flatten()
    
    return (x_train, y_train, x_test, y_test)

def scale_data(x_train, x_test):
    min_max_scaler = MinMaxScaler()
    x_train = min_max_scaler.fit_transform(x_train)
    x_test = min_max_scaler.transform(x_test)
    
    return (x_train, x_test)
    
def rotate(x, y, angle):
    angle = 10

    rad = angle*3.1415926535/180;

    new_x = x * math.cos(rad) - y * math.sin(rad);
    new_y = y * math.cos(rad) + x * math.sin(rad);
    
    return (new_x, new_y)

def translate(x, y, x_dist, y_dist):
    new_x = x + x_dist
    new_y = y + y_dist
    
    return (new_x, new_y)
    
def scale(x, y, factor_x, factor_y):
    new_x = x * factor_x
    new_y = y * factor_y
    
    return (new_x, new_y)

def shear(x, y, h_x, h_y):
    new_x = x + h_x * y
    new_y = y + h_y * x
    
    return (new_x, new_y)
```

```python
from random import randint

def random_transforms(feat_row):
    feat_row = feat_row.copy()
    x, y = split_x_y(feat_row)
    
    do_scale = rand_bool(0.6)
    do_scale_shake = rand_bool(0.6)
    do_rot = rand_bool(0.7)
    do_translate = rand_bool(0.7)
    do_shear = rand_bool(0.6)
    do_drop = rand_bool(0.7)
    
    if do_scale:
        scale_factr_x = random.uniform(0.7, 1.6)
        scale_factr_y = random.uniform(0.7, 1.6)
        for i in range(0, x.size):
            if x[i] != -1.0:
                new_xy = scale(x[i], y[i], scale_factr_x, scale_factr_y)
                x[i] = new_xy[0]
                y[i] = new_xy[1]
                
    if do_scale_shake:
        scale_factr_x = random.uniform(0.8, 1.2)
        scale_factr_y = random.uniform(0.8, 1.2)
        shake = 0.03
        for i in range(0, x.size):
            if x[i] != -1.0:
                new_xy = scale(x[i], y[i], scale_factr_x + random.uniform(-shake, shake), 
                               scale_factr_y + random.uniform(-shake, shake))
                x[i] = new_xy[0]
                y[i] = new_xy[1]
            
    if do_rot:
        angle = randint(-13, 13)
        for i in range(0, x.size):
            if x[i] != -1.0:
                new_xy = rotate(x[i], y[i], 10)
                x[i] = new_xy[0]
                y[i] = new_xy[1]
            
    if do_translate:
        displace = randint(-7, 7)
        for i in range(0, x.size):
            if x[i] != -1.0:   
                new_xy = translate(x[i], y[i], displace, displace)
                x[i] = new_xy[0]
                y[i] = new_xy[1]
            
    if do_shear:
        shear_factr = random.uniform(-0.10, 0.10)
        for i in range(0, x.size):
            if x[i] != -1.0:
                new_xy = shear(x[i], y[i], shear_factr, shear_factr)
                x[i] = new_xy[0]
                y[i] = new_xy[1]
                
    # drop points from beginning and end
    if do_drop:
        # don't drop when very few points
        if x.size > 8:
            nr_drop_start = randint(0, 2)
            nr_drop_end = randint(0, 2)
            
            new_x = drop_from_end(x, nr_drop_end)
            x = drop_from_start(new_x, nr_drop_start)
            new_y = drop_from_end(y, nr_drop_end)
            y = drop_from_start(new_y, nr_drop_start)
        
            
    new_row = np.zeros(x.size * 2)
    for i in range(0, x.size):
            new_row[2*i] = x[i]
            new_row[(2*i)+1] = y[i]
            
    return new_row

# True with probability
def rand_bool(probability):
    return random.random() < probability

def drop_from_end(row, drop_nr):
    if drop_nr > 0:
        nr_dropped = 0
        for i in range(row.size-1, 0, -1):
            if row[i] != -1.0:
                row[i] = -1.0
                nr_dropped += 1
                if nr_dropped == drop_nr:
                    break
    
    return row

def drop_from_start(row, drop_nr):
    if drop_nr > 0:
        row = row[drop_nr:]
        appendage = np.zeros(drop_nr)
        appendage.fill(-1.0)

        return np.hstack((row, appendage))
    else:
        return row
```

```python
def augment_data(feats, labels, nr_transforms=15):
    nr_rows = feats.shape[0]
    all_new_rows = []
    new_labels = []
    
    for i in range(0, nr_rows):
        row = feats[i,:]
        all_new_rows.append(row)
        new_labels.append(labels[i])
        
        # create transformed samples
        for j in range(0, nr_transforms):
            row = feats[i,:]
            new_row = random_transforms(row)
            
            all_new_rows.append(new_row)
            new_labels.append(labels[i])
    
    new_feats = np.vstack(all_new_rows)
    return (new_feats, np.asarray(new_labels))

def shuffle_train_data(feats, labels):
    all_data = np.column_stack((feats, labels))
    np.random.shuffle(all_data)
    shuf_feats = all_data[:, :-1]
    shuf_labels = all_data[:, -1]
    
    return (shuf_feats, shuf_labels)
```

```python
from __future__ import print_function
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D, LSTM
from keras import backend as K

def plot_history(histories):
    f, axarr = plt.subplots(2, len(histories))
    f.set_size_inches(10 * len(histories),20)

    for j in range(0, len(histories)):
        axarr[0, j].plot(histories[j].history['acc'])
        axarr[0, j].plot(histories[j].history['val_acc'])
        axarr[0, j].set_title('model accuracy')
        axarr[0, j].set_ylabel('accuracy')
        axarr[0, j].set_xlabel('epoch')
        axarr[0, j].legend(['train', 'test'], loc='upper left')
        
        #loss = summation of errors made
        axarr[1, j].plot(histories[j].history['loss'])
        axarr[1, j].plot(histories[j].history['val_loss'])
        axarr[1, j].set_title('model loss')
        axarr[1, j].set_ylabel('loss')
        axarr[1, j].set_xlabel('epoch')
        axarr[1, j].legend(['train', 'test'], loc='upper left')
        
    plt.show()
    


def train_test_model(x_train, y_train, x_test, y_test):

    # reshape input to be [samples, time steps, features]
    x_train = numpy.reshape(x_train, (x_train.shape[0], 1, x_train.shape[1]))
    x_test = numpy.reshape(x_test, (x_test.shape[0], 1, x_test.shape[1]))
    
    batch_size = 64
    num_classes = 10
    epochs = 20
    input_size = 64

    y_train = keras.utils.to_categorical(y_train, num_classes)
    y_test = keras.utils.to_categorical(y_test, num_classes)

    model = Sequential()
    #model.add(Dense(64, activation='relu', input_dim= input_size))
    model.add(LSTM(64, input_shape=(None, input_size)))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(num_classes, activation='softmax'))

    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0),
                  metrics=['accuracy'])

    history = model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=0,
          validation_data=(x_test, y_test))

    score = model.evaluate(x_test, y_test, verbose=0)

    print('Test loss:', score[0])
    print('Test accuracy:', score[1])
    
    return history
    
def train_test_with_folds(nr_folds=3):
    histories = []
    for _ in range(0, nr_folds):
        x_train, y_train, x_test, y_test = split_data(df, test_perc=0.35)
        new_feats, new_labels = augment_data(x_train, y_train, nr_transforms=15)
        x_train, y_train = shuffle_train_data(new_feats, new_labels)
        x_train, x_test = scale_data(x_train, x_test)

        history = train_test_model(x_train, y_train, x_test, y_test)
        histories.append(history)
    plot_history(histories)
    
train_test_with_folds()
```
